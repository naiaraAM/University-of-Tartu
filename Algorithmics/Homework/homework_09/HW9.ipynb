### Greedy analysis

A greedy algorithm can be defined as a really optimistig algorithm, due to its assumption that the next optimal step will lead to the optimal solution. It does not consider the whole overview of a problem and it is easy to make it fail. Besides, once the decision is made, it cannot be revesersed.

On easy example is going through a maze. A greedy algorithm can go any direction, but once a step is made i cannot be undo. If the step taken leads to no solution, it will stop and say that there is no solution, even if it was possible to solve the problem. It is true that greedy algorithms FIND an optimal solution but not always.

I create a simple example of searching the most weighted path from a node. The graph is a tree. I used for the optimal solution DFS. I represented both outcomes in a GIF (sorry for the infinite loop). The number that is not really well seen is 1000.

In the case of the greedy algorithm is easy to make it fail, by puting the weigth of one of the child nodes big enough so it selects it (in my case big enough is one unit bigger). Then if the children nodes of the non selected one are bigger by just 2 units, the algorithm will not find the optimal solution, as it is the case.
